```python
import requests
from bs4 import BeautifulSoup
import pandas
import json

# Problem statement 1
# please write the input separated by commas(,) without spaces, even if the input is give with spaces after commas
# the code will print first stock the will allow the user to inputs remaining stock one by one
# the code automatically converts lower case in uppercase
# if a number is entered as a input, the code allows user to input a correct value

# both pandas and json module is used in the code for preferred presentation of output
# code can be modified quickly to switch between pandas method (Tabular presentation)
# json format for as a whole presentation( personally prefer json)
# the written code take data from live stock market of nseindia.com to give the required output
# the site mention in the document of instruction new nseindia.com is not use, as I was not able to perform Webscraping
# due to no response from the site to request module

try:
    def func(stockname):
        while True:                  # to check for input if entered any number
            if stockname.isalpha():
                break
            else:
                stockname = (input("Enter a valid stock name: ")).upper()
        # stock url to be use
        stock_url = 'https://www.nseindia.com/live_market/dynaContent/live_watch/get_quote/GetQuote.jsp?symbol=' + str(
            stockname)
        # print(stock_url)
        headers = {
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'}

        # request module for gaining access to the content of the website
        response = requests.get(stock_url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        # fetch required section of the array
        data_array = soup.find(id='responseDiv').getText().strip().split(":")
        type(data_array)

        # following function contain the data needed
        # Some names of data may be change as per site showcase
        def data_required():

            def Lastest_Price():
                for item in data_array:
                    if 'lastPrice' in item:          # find the mention(in this case lastprice) data from the arrays
                        index = data_array.index(item) + 1   # stores the data
                        # print("Index -> "+ str(index))
                        latestPrice = data_array[index].split('"')[1]  # to get to the the data
                        return latestPrice
                        # print("Lastest Price: " + latestPrice)

            def Open_price():
                for item in data_array:
                    if 'open' in item:
                        index = data_array.index(item) + 1
                        # print("Index -> "+ str(index))
                        open = data_array[index].split('"')[1]
                        return open

            def Close_price():
                for item in data_array:
                    if 'close' in item:
                        index = data_array.index(item) + 1
                        # print("Index -> "+ str(index))
                        close = data_array[index].split('"')[1]
                        return close

            def Day_high():
                for item in data_array:
                    if 'dayHigh' in item:
                        index = data_array.index(item) + 1
                        # print("Index -> "+ str(index))
                        dayHigh = data_array[index].split('"')[1]
                        return dayHigh

            def Day_low():
                for item in data_array:
                    if 'dayLow' in item:
                        index = data_array.index(item) + 1
                        # print("Index -> "+ str(index))
                        dayLow = data_array[index].split('"')[1]
                        return dayLow

            def Volume():
                for item in data_array:
                    if 'totalTradedVolume' in item:
                        index = data_array.index(item) + 1
                        # print("Index -> "+ str(index))
                        totalTradedVolume = data_array[index].split('"')[1]
                        return totalTradedVolume

            def Delivery_quantity():
                for item in data_array:
                    if 'deliveryQuantity' in item:
                        index = data_array.index(item) + 1
                        # print("Index -> "+ str(index))
                        deliveryQuantity = data_array[index].split('"')[1]
                        return deliveryQuantity

            def Percent_deliver_quantity():
                for item in data_array:
                    if 'deliveryToTradedQuantity' in item:
                        index = data_array.index(item) + 1
                        # print("Index -> "+ str(index))
                        deliveryToTradedQuantity = data_array[index].split('"')[1]
                        return deliveryToTradedQuantity
            def required():
                l = []
                i = 2
                for i in range(1, 2):
                    d={}
                    d["Stock Name"] = stockname
                    d["Latest Price"] = Lastest_Price()
                    d["Open"] = Open_price()
                    d["close"] = Close_price()
                    d["Day High"] = Day_high()
                    d["Day Low"] = Day_low()
                    d["Quantity Traded(volume)"] = Volume()
                    d["Delivery Quantity"] = Delivery_quantity()
                    d["Delivery Percentage"] = Percent_deliver_quantity() + '%'
                    l.append(d)
                    # pandas.set_option('display.max_rows', None)
                    # pandas.set_option('display.max_columns', None)
                    # pandas.set_option('display.width', 2000)
                    # pandas.set_option('display.float_format', '{:20,.2f}'.format)
                    # pandas.set_option('display.max_colwidth', None)
                    # pandas.get_option('display.max_columns')
                    # pandas.set_option('display.max_columns',None)
                    #df = pandas.DataFrame(l)
                    json_object = json.dumps(d, indent= 4 )
                    print(json_object)




                # print(df)

            return required()
        return data_required()





        func(stockname)

except TypeError:
    print("Invalid Name")

Stock_name = list(map(str, input("Enter the name of stock or stocks, seperate the stocks by comma(,) without spaces"
                                 "\n (Example:RELIANCE,HDFCBANK,ADANIPORTS,ITC,SBIN,IOC,RBLBANK)   : ").split(",")))

e=0
for e in range(len(Stock_name)):
    stockname = Stock_name[e]
    e = e + 1
    func(stockname)
```
